```python
import numpy as np 

rev = [100, 1000, 2000.0, 4000, 3000, 200 , 345, 123, 34500]
rev_array = np.array(rev)

print(rev_array.dtype) # show the type of array 

rev_selec = rev_array[(rev_array>100) & (rev_array < 1000)]

print("rev selct = ", rev_selec )
```

Création de tableau 

```python
# crée un tableau de random comprise entre 0 et 1
rand01 = np.random.random((6,3)) 

print("tab of random value from 0 to 1 = ", rand01)

# Crée un tableau de random d'entier 
randint = np.random.randint(0, 10,size=(5,4))

print("tab of random integer = ", randint)

# ajouter une lingne à un tableau existant 
line = [0, 4, 6, 123]

randint = np.vstack((randint, line)) # see also hstack 

print("modified tab randint = ", randint)
```
